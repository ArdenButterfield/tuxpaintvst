```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
BASE_PERIOD = 32
SAWTOOTH = np.array([i / BASE_PERIOD for i in range(BASE_PERIOD)])
SAWTOOTH -= np.sum(SAWTOOTH) / len(SAWTOOTH)
BASE_WAVE = SAWTOOTH
```

```python
plt.plot(np.tile(BASE_WAVE, 5))
```

```python
transform = np.fft.fft(BASE_WAVE)
print(transform)
# transform = transform[:len(transform) // 2 + 1]
plt.scatter(transform[1:].real, transform[1:].imag)

mags = abs(transform)
phases = np.angle(transform)
# plt.plot(mags)
plt.plot(phases)
```

```python
a = np.tile(np.arange(BASE_PERIOD) * 2 * np.pi / BASE_PERIOD, BASE_PERIOD)
a = np.reshape(a, (BASE_PERIOD, BASE_PERIOD))
for i in range(BASE_PERIOD):
    a[i] = (mags[i] * np.cos((i * a[i]) + phases[i]) / BASE_PERIOD)# + (mags[i] * np.cos((len(transform) - i * a[i]) + phases[i]) / len(transform))
plt.imshow(a)
```

```python
a = np.tile(np.arange(BASE_PERIOD) * 2 * np.pi / BASE_PERIOD, BASE_PERIOD // 2 + 1)
a = np.reshape(a, (BASE_PERIOD // 2 + 1, BASE_PERIOD))
a[0] = 0 * a[0] + mags[0]
for i in range(1, BASE_PERIOD // 2):
    a[i] = 2 * (mags[i] * np.cos((i * a[i]) + phases[i]) / BASE_PERIOD)# + (mags[i] * np.cos((len(transform) - i * a[i]) + phases[i]) / len(transform))
i = BASE_PERIOD // 2
a[i] = (mags[i] * np.cos((i * a[i]) + phases[i]) / BASE_PERIOD)
plt.imshow(a)
```

```python
plt.plot(np.sum(a, axis=0))
```
